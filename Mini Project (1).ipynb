# CLV prediction 

```python
import os
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cluster import KMeans
```

```python
os.makedirs("models", exist_ok=True)
os.makedirs("outputs", exist_ok=True)
RANDOM_STATE = 42
```

```python
df['frequency'] = df['Total_Transactions']
df['monetary'] = df['Total_Spent']
df['value_freq_interaction'] = df['Avg_Transaction_Value'] * df['frequency']
df['cashback_spend_ratio'] = np.where(df['Total_Spent'] > 0,
                                      df['Cashback_Received'] / df['Total_Spent'], 0.0)
df['Retention_Rate'] = np.clip(np.random.normal(0.7, 0.1, size=len(df)), 0, 1)
```

```python
df['Future_CLV'] = df['Avg_Transaction_Value'] * df['Total_Transactions'] * df['Retention_Rate']
df['Future_CLV_log'] = np.log1p(df['Future_CLV'])
```

```python
feature_cols_num = [
    'Age','frequency','monetary','Avg_Transaction_Value','Max_Transaction_Value',
    'Min_Transaction_Value','Active_Days','Loyalty_Points_Earned','Referral_Count',
    'Cashback_Received','Support_Tickets_Raised','Issue_Resolution_Time',
    'Customer_Satisfaction_Score','value_freq_interaction','cashback_spend_ratio'
]

feature_cols_cat = ['Location','Income_Level','App_Usage_Frequency','Preferred_Payment_Method']

X = df[feature_cols_num + feature_cols_cat]
y_clv = df['Future_CLV_log']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_clv, test_size=0.2, random_state=RANDOM_STATE)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), feature_cols_num),
        ('cat', OneHotEncoder(handle_unknown='ignore'), feature_cols_cat)
    ]
)
```

```python
clv_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', GradientBoostingRegressor(random_state=RANDOM_STATE))
])

clv_params = {
    'regressor__n_estimators': [200, 300],
    'regressor__max_depth': [3, 5],
    'regressor__learning_rate': [0.05, 0.1]
}

clv_grid = GridSearchCV(clv_pipeline, clv_params, cv=3,
                        scoring='neg_root_mean_squared_error', n_jobs=-1, verbose=1)
clv_grid.fit(X_train, y_train)
```

```python
joblib.dump(clv_grid.best_estimator_, "models/clv_model.pkl")
print(" Saved CLV model -> models/clv_model.pkl")
```

```python
y_pred_log = clv_grid.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_true = np.expm1(y_test)

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)
mape = np.mean(np.abs((y_true - y_pred)/y_true))*100

print(" CLV RMSE:", rmse)
print(" CLV R2:", r2)
print(" CLV MAPE (%):", mape)
```

```python
df.columns.tolist()
```

```python
# 1 Predict Future CLV for full dataset
df['Future_CLV_Pred'] = np.expm1(clv_grid.predict(X))

# 2 Cluster customers based on CLV and other metrics
seg_features = ['Future_CLV_Pred', 'Total_Transactions', 'Total_Spent', 'value_freq_interaction']
kmeans = KMeans(n_clusters=3, random_state=RANDOM_STATE)
df['Customer_Segment'] = kmeans.fit_predict(df[seg_features])

# Map segment numbers to labels
seg_order = df.groupby('Customer_Segment')['Future_CLV_Pred'].mean().sort_values(ascending=False).index
seg_map = {seg_order[0]: 'High Value', seg_order[1]: 'Medium Value', seg_order[2]: 'Low Value'}
df['Customer_Segment'] = df['Customer_Segment'].map(seg_map)

# 3 Then assign business actions
def business_action(row):
    if row['Customer_Segment'] == 'High Value':
        return "Upsell: premium/VIP services"
    elif row['Customer_Segment'] == 'Medium Value':
        return "Engagement: personalized nudges"
    else:
        return "Standard nurturing"

df['Business_Action'] = df.apply(business_action, axis=1)
```

```python
regressor = clv_grid.best_estimator_['regressor']
num_features = feature_cols_num
cat_features = list(clv_grid.best_estimator_['preprocessor'].transformers_[1][1].get_feature_names_out())
features = num_features + cat_features
importances = regressor.feature_importances_

fi_df = pd.DataFrame({'Feature': features, 'Importance': importances}).sort_values(by='Importance', ascending=False).head(20)
plt.figure(figsize=(10,6))
sns.barplot(x='Importance', y='Feature', data=fi_df)
plt.title("Top 20 Features: CLV Model")
plt.tight_layout()
plt.show()
```
